# IMPORTING AND CLEANING DATA

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head()
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head()
```

```python
print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts())
print(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts())
```

```python
yes_no_dict = {
    "Yes": True,
    "No": False
}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no_dict)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no_dict)

star_wars.head()
```

```python
import numpy as np

movie_dict = {
    "Star Wars: Episode I  The Phantom Menace":True,
    "Star Wars: Episode II  Attack of the Clones":True,
    "Star Wars: Episode III  Revenge of the Sith":True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True,
    np.nan:False
}

for i in range(3,9):
    star_wars.iloc[:,i] = star_wars.iloc[:,i].map(movie_dict)

star_wars.head()
```

```python
name_dict = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': "seen_1",
    'Unnamed: 4': "seen_2",
    'Unnamed: 5': "seen_3",
    'Unnamed: 6': "seen_4",
    'Unnamed: 7': "seen_5",
    'Unnamed: 8': "seen_6",
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "ranking_1",
    'Unnamed: 10': "ranking_2",
    'Unnamed: 11': "ranking_3",
    'Unnamed: 12': "ranking_4",
    'Unnamed: 13': "ranking_5",
    'Unnamed: 14': "ranking_6",
}

star_wars = star_wars.rename(columns=name_dict)

star_wars.head()
```

```python
star_wars.iloc[:,9:15] = star_wars.iloc[:,9:15].astype(float)

star_wars.head()
```

```python
%matplotlib inline
star_wars.iloc[:,9:15].mean().plot.bar()
```

# EXPLORING DIFFERENCES BETWEEN FANS AND NON-FANS

```python
import matplotlib.pyplot as plt

fans_col = "Do you consider yourself to be a fan of the Star Wars film franchise?"
fans = star_wars[star_wars[fans_col] == True]
not_fans = star_wars[star_wars[fans_col] == False]

fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(121)
fans.loc[:,"ranking_1":"ranking_6"].mean().plot.bar()
ax.legend(["Fans"])

ax = fig.add_subplot(122)
not_fans.loc[:,"ranking_1":"ranking_6"].mean().plot.bar()
ax.legend(["Not Fans"])

fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(121)
fans.loc[:,"seen_1":"seen_6"].sum().plot.bar()
ax.legend(["Fans"])

ax = fig.add_subplot(122)
not_fans.loc[:,"seen_1":"seen_6"].sum().plot.bar()
ax.legend(["Not Fans"])
```

## Analysis on Fans x Non-fans differences

There are two significant differences between those who declares themselves as fans of Star Wars and those who don't:

- 1: The fans tend to watch all the movies equally. That is: even though they have their preferences, they won't avoid watching a movie because of that. On the other side, non-fans don't bother watching movies that are less popular.
- 2: Fans tend to rank the first trilogy to be released (Episodes from IV to VI) considerably better than the second trilogy (Episodes from I to III). Non-fans tends to have a less biased ranking regarded to which trilogy the movie belongs to (there are movies in the second trilogy that perform better than movies in the frist trilogy - and there is a technical tie between which movie is the best).
